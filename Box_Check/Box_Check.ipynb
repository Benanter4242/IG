```python
import pandas as pd
```

```python
# Read the CSV files
US = pd.read_csv("/content/drive/MyDrive/IG /Box_Check/Filebank.csv")
Warehouse = pd.read_csv("/content/drive/MyDrive/IG /Box_Check/Warehouse.csv")
```

```python
# Initialize lists and dictionary for processing
US_box = []
Warebox = []
Warefiles = {}
```

```python
# Process the US DataFrame
for index, row in US.iterrows():
    # Split the files by whitespace and strip any leading/trailing spaces
    Files = row["Files in boxes"].strip().split()
    # Strip any leading/trailing spaces from the box name
    Box = row["Box"].strip()
    # Append the box and sorted list of files to the US_box list
    US_box.append([Box, sorted(Files)])
```

```python
# Create a DataFrame from the processed US box data
US_box_df = pd.DataFrame(US_box, columns=["Box", "Our Files"])
```

```python
# Process the Warehouse DataFrame
for index, row in Warehouse.iterrows():
    # Strip any leading/trailing spaces from the box and file names
    WBox = row["Box"].strip()
    WFile = row["File"].strip()
    # Append the file to the list in the dictionary for the corresponding box
    if WBox not in Warefiles:
        Warefiles[WBox] = WFile
    else:
        Warefiles[WBox] += ',' + WFile
```

```python
# Convert the dictionary to a list of tuples and then to a DataFrame
Warefiles_list = [(key, value) for key, value in Warefiles.items()]
Warefiles_df = pd.DataFrame(Warefiles_list, columns=['Box', 'WFiles'])
```

```python
# Process the Warehouse file data
for index, row in Warefiles_df.iterrows():
    # Split the files by comma and strip any leading/trailing spaces
    WareF = row["WFiles"].strip().split(',')
    # Strip any leading/trailing spaces from the box name
    WareB = row["Box"].strip()
    # Append the box and sorted list of files to the Warebox list
    Warebox.append([WareB, sorted(WareF)])
```

```python
# Create a DataFrame from the processed Warehouse box data
Warebox_df = pd.DataFrame(Warebox, columns=["Box", "WFiles"])

# Merge the US box DataFrame with the Warehouse box DataFrame
research_pd = pd.merge(Warebox_df, US_box_df, how="right", on='Box')

research_pd["Same?"]= research_pd["WFiles"] == research_pd["Our Files"]

research_pd["WFiles"]= research_pd["WFiles"].fillna("File not found")
```

```python
# Save the result to a CSV file
research_pd.to_csv("research.csv", index=False)

# Print the resulting DataFrame
print(research_pd)
```
